```python
import pandas as pd
data=pd.read_csv('/content/company-sales.csv')
data.head()
```

```python
print('approch1:keep the missing value as is')
print(data.isnull().sum())
print()
```

```python
print('aproch2:remove data object with missing value')
column_row=data.dropna()
print(column_row)
print()
```

```python
print('aproch3:remove attribute with missing values')
column_cleaned=data.dropna(axis=1)
print(column_cleaned)
print()
```

```python
print('approch4:estimate and impute with missing value')
data_impute_mean=data.fillna(data.mean())
print(data_impute_mean)
data_interpolation=data.interpolate()
print(data_interpolation)
print()
```
